```python
df = pd.read_csv('Global YouTube Statistics.csv', encoding='ISO-8859-1')
df.head(10)
```

```python
media = df['subscribers'].head(10).mean()


print(f'A média de inscritos dos 10 canais com mais incritos do youtube é: {media}')
```

```python
mediana = df['subscribers'].head(10).median()


print(f'A mediana de inscritos dos 10 canais com mais incritos do youtube é: {mediana}')
```

```python
variancia = df['subscribers'].head(10).var()

print(f'A variancia de inscritos dos 10 canais com mais incritos do youtube é: {variancia}')
```

```python
desvio_padrao = df['subscribers'].head(10).std()

print(f'O devio padrão de inscritos dos 10 canais com mais incritos do youtube é: {desvio_padrao}')
```

```python
# Histograma
plt.hist(df['subscribers'], bins=20, edgecolor='k')
plt.xlabel('Número de Inscritos (em 100 milhoes)')
plt.ylabel('Frequência')
plt.title('Histograma do Número de Inscritos')

plt.show()
```

```python
# Aqui estão contidos todos os 1000 canais com mais inscritos
# Gráfico de dispersão
plt.scatter(df['subscribers'], df['uploads'])
plt.xlabel('Número de Inscritos (em 100 milhoes)')
plt.ylabel('Número de Uploads')
plt.title('Gráfico de Dispersão entre Número de Inscritos e Número de Uploads')
plt.show()
```
